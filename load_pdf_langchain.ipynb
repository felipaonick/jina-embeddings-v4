# Load a pdf file using LangChain 

```python
# -------------------------------------------
# 1Ô∏è‚É£ Importa librerie
# -------------------------------------------
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from transformers import AutoTokenizer, AutoModel
import torch

# -------------------------------------------
# 2Ô∏è‚É£ Carica PDF e splitta in chunk
# -------------------------------------------

# üìÑ Percorso PDF
pdf_path = "./414759-1-_5_Nike-NPS-Combo_Form-10-K_WR.pdf"

# üîπ Usa PyPDFLoader
loader = PyPDFLoader(pdf_path)
pages = loader.load()

# üîπ Splitta con RecursiveCharacterTextSplitter
splitter = RecursiveCharacterTextSplitter(
    chunk_size=1500,
    chunk_overlap=200,
    separators=["\n\n", "\n", " ", ""]
)

chunks = splitter.split_documents(pages)

# üîπ Estrai solo testo dei chunk
texts = [chunk.page_content for chunk in chunks]

print(f"Numero chunk: {len(texts)}")
print(texts[:2])  # mostro primi due chunk
```
